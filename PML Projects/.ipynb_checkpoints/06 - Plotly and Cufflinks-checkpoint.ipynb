# Plotly and Cufflinks

* interactive visualization library
* connects with plotly and pandas
* must be installed
* pip install plotly
* pip install cufflinks


```python
import pandas as pd 
import numpy as np 
from plotly import __version__
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
%matplotlib inline
```

```python
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
#data
df = pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'category':['A','B','C'], 'Values':[32,43,50]})
```

```python
df2
```

```python
df.plot()
```

```python
df.iplot()
```

```python
#scatter plots (dont forget to add markers)
df.iplot(kind='scatter', x= 'A',y='B', mode='markers', size=5)
```

```python
df2
```

```python
df2.iplot(kind='bar',x='category', y ='Values')
```

```python
df.head()
```

```python
#if we did a bar plot as is...
df.iplot(kind='bar')
#not very useful...
```

```python
df.count().iplot(kind='bar')
```

```python
df.sum().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

```python
# 3d surface plots  
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
```

```python
df3
```

```python
df3.iplot(kind='surface',colorscale='rdylbu')
```

```python
df['A'].iplot(kind='hist',bins =50)
```

```python
df.iplot(kind='hist',bins =50)
```

```python
df[['A','B']].iplot(kind='spread')
```

```python
df.iplot(kind='bubble',x='A',y='B', size='C')
```

```python
#careful when using large datasets will likely crash
df.scatter_matrix()
```
