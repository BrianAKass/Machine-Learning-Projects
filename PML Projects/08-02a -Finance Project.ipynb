```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime 
%matplotlib inline
```

```python
#bankdata file in case we need it due to API issues
df = pd.read_pickle('all_banks')

start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
```

```python
#google api can't be reached reliably have to use yahoo 
BAC = data.DataReader('BAC', 'yahoo', start, end)
C = data.DataReader('C', 'yahoo', start, end)
GS= data.DataReader('GS', 'yahoo', start, end)
JPM= data.DataReader('JPM', 'yahoo', start, end)
MS= data.DataReader('BAC', 'yahoo', start, end)
WFC= data.DataReader('BAC', 'yahoo', start, end)
```

```python
BAC = BAC.drop('Adj Close',axis=1)
C = C.drop('Adj Close',axis=1)
GS =GS.drop('Adj Close',axis=1)
JPM =JPM.drop('Adj Close',axis=1)
MS = MS.drop('Adj Close',axis=1)
WFC =WFC.drop('Adj Close',axis=1) 
```

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
stockFrames=[BAC,C,GS,JPM,MS,WFC]
```

```python
bank_stocks = pd.concat(stockFrames, axis=1, keys=tickers)
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
bank_stocks.head()
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

```python
returns = pd.DataFrame()
```

```python
for x in tickers:
    returns[x+' return'] = bank_stocks[x]['Close'].pct_change()
returns.head()
```
