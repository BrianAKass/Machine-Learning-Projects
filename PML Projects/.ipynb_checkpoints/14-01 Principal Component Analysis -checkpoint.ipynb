# Principal Component Analysis

* Unsupervised statistical technique used to examine the interrelations amoung a set of variables in order to identify the underlying structure of those variables.
* It is also known sometimes as a general factor analysis.

* Where regression dtermines a line of best fitto a data set, factor analysis determines several orthogonal lines of best  fit to data set. 

* Orthognal meaning "at right angles" 
    * actually the lines are perpendicular to each other in n-dimensional space. 
* n-Dimensional Space is the variable sample space. 
    * There is as many dimensions as there are variables, so in a data set with 4 variables the sample space is 4-dimensional.
    
* Here we have some data plotted along two features, x and y.


* Now we can begin to understand the componenents!
* Componenets are linear transformation that chooses a variable system for the data set such that the greatest variance of the data set comes to lie on the first axis. 



* The second greatest variance on the second axis, and so on...



* This process allows us to reduce the number of variables used in an analysis.

* Note that the components are uncorrelated, since in the sample space they are orthagonal to each other. 



* we can contribute this analysis to higher dimensions

* If we use this technique ona data set with a large number of variables, we can compress the amount of explained variationto just a few components

* The most challenging part of PCA is interpreting the components.
