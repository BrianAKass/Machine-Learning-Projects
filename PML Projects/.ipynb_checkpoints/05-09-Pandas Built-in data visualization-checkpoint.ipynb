```python
#create a new dataframe using random numbers generating a 1000 * 2, 
# two cols are named a and b
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
df.head()
```

```python
#hexBIN hexbin hexbin hexbin hexbin
#gridsize less = bigger
df.plot.hexbin(x='a',y='b',gridsize=25)
```

```python
#don't forgert useful args like edgecolor and lw

df.plot.hexbin(x='a',y='b',gridsize=25,cmap='plasma',edgecolor='w', lw=0.001)
```
