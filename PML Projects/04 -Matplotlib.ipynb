# Matplotlib 
* Most popular plotting library for python
* gives you control over every aspect of a figure
* designed to have a similar feel to Matlib's graphical plotting.

* You'll need to install it with pip or conda at your command line or terminal with: 
    * conda install matplotlib
    * or 
    * pip install matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
## OO
fig  = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

plt.show()
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
plt.show()
```

```python
fig,axes  = plt.subplots(nrows=1,ncols=2)
# axes.plot(x,y)
# plt.show(x,y)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('Second plot')
plt.tight_layout()
plt.show()
```

# Fig Size and DPI

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
plt.show()
```

```python
fig.savefig('my_picture.png',dpi=3000)
```

```python
fig=plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('title')
ax.set_xlabel('XXXXXXXX')
ax.set_ylabel('YYYYYYYY')


plt.tight_layout()
plt.show()
```

```python
fig=plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label= 'Blue')
ax.plot(x,x**3, label= 'Yellow')

ax.legend()

plt.tight_layout()
plt.show()
```

```python
fig=plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label= 'Blue')
ax.plot(x,x**3, label= 'Yellow')

ax.legend(loc=0)

plt.tight_layout()
plt.show()
```

# Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, c='red',lw = 3, alpha =0.5 ) #hex codes work as well

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, c='red',lw = 3, ls='-.' ) #ls =linestyle : - . steps

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, c='red',lw = 1,marker = 'o', markersize=10,
       mfc= 'white',mew = 1,mec='red') 
#marker o . + x makerfacecolor = mfc 
#markeredgewidth = mew 
#markeredgecolor = mec
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, c='red',lw = 2,ls = '--') 
#configure the range of the axes x and y for tightly fitted axes range
ax.set_xlim([0,1])
ax.set_ylim([0,2])
plt.tight_layout()
plt.show()
```
