```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import cufflinks as cf
sns.set_style('whitegrid')
%matplotlib inline
cf.go_offline()
```

```python
#in case Datareader fails to load google API and other data gatehring API
df = pd.read_pickle('all_banks')
```

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
```

```python
##google api can't be reached reliably have to use yahoo 
BAC = data.DataReader('BAC', 'yahoo', start, end)
C = data.DataReader('C', 'yahoo', start, end)
GS= data.DataReader('GS', 'yahoo', start, end)
JPM= data.DataReader('JPM', 'yahoo', start, end)
MS= data.DataReader('BAC', 'yahoo', start, end)
WFC= data.DataReader('BAC', 'yahoo', start, end)
##drop Adj Close col we don't need
BAC = BAC.drop('Adj Close',axis=1)
C = C.drop('Adj Close',axis=1)
GS =GS.drop('Adj Close',axis=1)
JPM =JPM.drop('Adj Close',axis=1)
MS = MS.drop('Adj Close',axis=1)
WFC =WFC.drop('Adj Close',axis=1) 
```

```python
tickers = 'BAC C GS JPM MS WFC'.split(' ')
stocksNames = [BAC,C,GS,JPM,MS,WFC]
tickers
```

```python
bank_stocks = pd.concat(stocksNames,keys=tickers,axis=1)
bank_stocks.columns.names= ['Bank Ticker', 'Stock Info']
bank_stocks.head()
```
