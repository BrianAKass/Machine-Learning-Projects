```python
import numpy as np
%matplotlib inline
x = np.arange(0,100)
y = x*2
z = x**2
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.2,.2])

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.2,.2])
axes1.plot(x,y,c='r')
axes2.plot(x,y,c='r')
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,0.4,0.4])
axes1.set_xlabel('X',fontsize=14)
axes1.set_ylabel('Z',fontsize=14)
axes1.set_title('Programming  Knowledge Over Time',fontsize=14)
axes1.plot(x,z)
axes2.plot(x,y)
axes2.set_xlabel('X',fontsize=10)
axes2.set_ylabel('Y',fontsize=10)
axes2.set_title('Zoom',fontsize=12)


plt.tight_layout()
plt.show()
```

```python
fig,axes= plt.subplots(nrows=1,ncols=2) 
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,c='b',lw=3,ls='--')
axes[1].plot(x,z,c='r',lw=3,ls='-')

plt.tight_layout()
plt.show()
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(10,2))
ax[0].plot(x,y,c='b',lw=2,ls='--')
ax[1].plot(x,z,c='r',lw=2)
plt.tight_layout()
plt.show()
```
