# Categorical Plots

```python
import seaborn as sns
import numpy  as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

# Quick explanation for reading box plots
#### https://www.youtube.com/watch?v=CoVf1jLxgj4

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data = tips)
```

```python
sns.violinplot(x='day',y='total_bill', data = tips, hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill', data = tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data = tips) 
#jitter is true by default
```

```python
sns.stripplot(x='day',y='total_bill',data =tips, jitter =False)
```

```python
sns.stripplot(x='day',y='total_bill',data = tips, hue='sex', dodge =True)
```

```python
# when violin and strip plots work together
sns.swarmplot(x='day',y='total_bill',data = tips)
```

```python
sns.swarmplot(x='day',y='total_bill',data = tips, hue= 'sex')
```

# Tip : 
## Don't use swarm plots for large datasets. very taxxing on the CPU

```python
sns.violinplot(x='day',y='total_bill',data = tips)
sns.swarmplot(x='day',y='total_bill',data = tips, color= 'black')
```

```python
# sns.factorplot(x='day', y= 'total_bill', data = tips)
# factorplot is now catplot as in category 
# stick with convention don't bother with kind unless you are switching  
# kinds of plots on the same cell for someone over your shoulder.

sns.catplot(x='day', y= 'total_bill', data = tips, kind='bar')
```

```python
sns.catplot(x='day', y= 'total_bill', data = tips, kind='box')
```

```python
sns.catplot(x='day', y= 'total_bill', data = tips, kind='violin')
```
